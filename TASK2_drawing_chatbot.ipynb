

*   Importing all the packages




```python
import pytesseract
import numpy as np
import gradio as gr
import os
import fitz  # For PDF processing
import torch
import faiss  # For vector search
from PIL import Image
from pytesseract import image_to_string  # For OCR
from sympy import sympify  # For mathematical parsing
from transformers import CLIPProcessor, CLIPModel, AutoTokenizer, AutoModelForCausalLM
```



*   This step sets up the machine learning models required for the project. Here's an explanation of each component:





```python
# Initialize Models
clip_model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14")
rag_generator_tokenizer = AutoTokenizer.from_pretrained("facebook/opt-125m")
rag_generator = AutoModelForCausalLM.from_pretrained("facebook/opt-125m")
```



*   This function is responsible for extracting images from a PDF file and saving them as separate image files.




```python
# Step 1: Extract Images from PDF
def extract_images_from_pdf(pdf_path, output_dir):
    doc = fitz.open(pdf_path)
    image_paths = []
    for page_num in range(len(doc)):
        for img_index, img in enumerate(doc[page_num].get_images(full=True)):
            xref = img[0]
            base_image = doc.extract_image(xref)
            img_bytes = base_image["image"]
            img_path = os.path.join(output_dir, f"page_{page_num + 1}_img_{img_index + 1}.png")
            with open(img_path, "wb") as img_file:
                img_file.write(img_bytes)
            image_paths.append(img_path)
    return image_paths
```



*    Embed Images using CLIP
This step involves converting extracted images into vector embeddings using the CLIP (Contrastive Language-Image Pretraining) model. These embeddings are numerical representations of the images, which allow for similarity-based search and further analysis.List item



```python
# Step 2: Embed Images using CLIP
def embed_images(image_paths):
    embeddings = {}
    for img_path in image_paths:
        image = Image.open(img_path).convert("RGB")
        inputs = clip_processor(images=image, return_tensors="pt")
        outputs = clip_model.get_image_features(**inputs).detach().numpy()
        embeddings[img_path] = outputs
    return embeddings
```



*   This step involves building a FAISS (Facebook AI Similarity Search) index to efficiently perform similarity searches over the image embeddings generated in Step 2. FAISS is a library optimized for fast nearest neighbor search in high-dimensional spaces.




```python
# Step 3: Build FAISS Index
def build_faiss_index(image_embeddings):
    dimension = list(image_embeddings.values())[0].shape[-1]
    index = faiss.IndexFlatL2(dimension)
    paths = list(image_embeddings.keys())
    vectors = np.vstack(list(image_embeddings.values()))
    index.add(vectors)
    return index, paths
```



*   This step is responsible for performing the similarity search to find the closest image based on the query embedding. Using the FAISS index built in Step 3, the function searches for the most relevant image for a given query.




```python
# Step 4: Retrieve Best Match
def retrieve_best_match(index, paths, query_embedding):
    _, indices = index.search(query_embedding, 1)
    return paths[indices[0][0]] if indices[0][0] != -1 else None
```



*  The process_image_for_math function uses OCR (Optical Character Recognition) to extract mathematical expressions from an image, processes them, and evaluates the results. If any errors occur during this process, it gracefully handles them and provides appropriate feedback.




```python
from pytesseract import TesseractNotFoundError, TesseractError

def process_image_for_math(image_path):
    try:
        # Extract text using OCR
        image = Image.open(image_path)
        ocr_text = pytesseract.image_to_string(image)

        # Extract and evaluate mathematical expressions
        expressions = []
        for line in ocr_text.splitlines():
            try:
                expr = sympify(line)  # Parse the line as a mathematical expression
                result = expr.evalf()  # Evaluate the expression
                expressions.append((line, result))  # Store as a tuple
            except:
                continue  # Skip non-mathematical lines

        return expressions
    except TesseractNotFoundError:
        return [("Error: Tesseract is not installed or not found in PATH.", None)]
    except TesseractError as e:
        return [(f"Tesseract error: {e}", None)]
    except Exception as e:
        return [(f"Unexpected error: {e}", None)]

```



*  The generate_response function is designed to generate an answer to the user's query by combining image matching, Optical Character Recognition (OCR), and mathematical expression evaluation. It performs a sequence of steps to ensure that the response contains relevant information from the matched image, especially if mathematical content is present in the image.




```python
# Step 6: Generate Response
def generate_response(question, matched_image_path):
    if not matched_image_path:
        return "No relevant image found for the query."

    # Perform OCR and math processing
    math_results = process_image_for_math(matched_image_path)

    if math_results:
        response = f"The image '{matched_image_path}' contains the following results:\n"
        for item in math_results:
            if len(item) == 2:  # Ensure item is a tuple with two elements
                expr, result = item
                response += f"Expression: {expr} = {result}\n"
            else:
                response += f"Unrecognized data: {item}\n"
        return response
    else:
        context = f"The image '{matched_image_path}' matches the question: {question}."
        inputs = rag_generator_tokenizer.encode(context, return_tensors="pt")
        outputs = rag_generator.generate(inputs, max_length=100)
        return rag_generator_tokenizer.decode(outputs[0], skip_special_tokens=True)
```



*   The chatbot_response function is designed to handle the entire process of generating a response to a user's question, including embedding the question, finding the most relevant image, and then generating a response based on the image conten




```python
# Chatbot Logic
def chatbot_response(user_question):
    # Embed the question using CLIP
    inputs = clip_processor(text=user_question, return_tensors="pt")
    query_embedding = clip_model.get_text_features(**inputs).detach().numpy()

    # Retrieve best match
    matched_image_path = retrieve_best_match(faiss_index, image_paths_list, query_embedding)

    # Generate response
    response = generate_response(user_question, matched_image_path)
    return f"Matched Image: {matched_image_path}\n\nResponse: {response}"
```



*   The Main Execution section of the code performs the entire process of extracting images from a PDF, embedding the images, building a FAISS index, and then using that to generate a response based on a user's query




```python
# Main Execution
if __name__ == "__main__":
    # PDF Path and Extraction Directory
    pdf_path = "/content/Autodesk Part Drawings.pdf"
    output_dir = "./extracted_images"
    os.makedirs(output_dir, exist_ok=True)

    print("Extracting images...")
    image_paths = extract_images_from_pdf(pdf_path, output_dir)

    print("Embedding images...")
    image_embeddings = embed_images(image_paths)

    print("Building FAISS index...")
    faiss_index, image_paths_list = build_faiss_index(image_embeddings)

    # Example query from user
    user_query = "Calculate the values in the table on page 2"

    print("Generating response...")
    response = chatbot_response(user_query)
    print(f"Response: {response}")
```



*  This part of the code initializes and launches the chatbot interface using the Gradio library, which creates an interactive web interface for the chatbot




```python
# Launch Chatbot
chatbot = gr.Interface(
    fn=chatbot_response,
    inputs="text",
    outputs="text",
    title="Engineering Drawing Math Chatbot",
    description="Ask questions about engineering drawings, and the chatbot will find the most relevant image, extract calculations, and provide an answer."
)
chatbot.launch(debug=True)
```
